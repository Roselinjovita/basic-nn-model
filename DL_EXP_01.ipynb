```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
from google.colab import auth
import gspread
from google.auth import default
import pandas as pd
```

```python
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
worksheet = gc.open('DEEP').sheet1
rows = worksheet.get_all_values()
df = pd.DataFrame(rows[1:], columns=rows[0])
dataset=df.astype({'input':'int'})
dataset=df.astype({'output':'int'})
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
x=dataset[['input']].values
```

```python
x
```

```python
y=dataset[['output']].values
```

```python
y
```

```python

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=33)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
scaler=MinMaxScaler()
scaler.fit(x_train)
```

```python
x_train1=scaler.transform(x_train)
```

```python
learn=Sequential([
    Dense(units=9,activation='relu',input_shape = [1]),
    Dense(units=9,activation='relu'),
    Dense(units=1)
])
```

```python
learn.compile(optimizer='rmsprop',loss='mse')
```

```python
learn.fit(x_train1,y_train,epochs=2000)
```

```python
loss_df=pd.DataFrame(learn.history.history)
```

```python
loss_df.plot()
```

```python
x_test1=scaler.transform(x_test)
```

```python
learn.evaluate(x_test1,y_test)
```
